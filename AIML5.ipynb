```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python
data=pd.read_csv("supermarket_sales.csv")
data.info()
```

```python
#1 line plot
plt.figure()
plt.plot(data['Unit price'],data['Total'], marker='*')
plt.title('Line plot')
plt.xlabel('Unit price')
plt.ylabel('Total')
plt.show()
```

```python
#2 Bar chart
plt.figure()
plt.bar(data['Quantity'], data['cogs'])
plt.title("Bar chart")
plt.xlabel('Quantity')
plt.ylabel('cogs')
plt.show()
```

```python
#3 Sactter plot
plt.figure()
plt.scatter(data['Quantity'], data['Unit price'])
plt.title('scatter plot')
plt.xlabel('Quantity')
plt.ylabel('Unit price')
plt.show()
```

```python
#4 Histogram
plt.figure()
sns.histplot(data['Quantity'], bins=6)
plt.title('Histogram')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#5 Box Plot
plt.figure()
sns.boxplot(x=data['Quantity'], y='Tax 5%', data=data)
plt.title('Box plot')
plt.xlabel('Quantity')
plt.ylabel('Tax 5%')
plt.show()
```

```python
#6 Heatmap (seaborn)
df=data[['Quantity', 'Unit price', 'Tax 5%', 'cogs', 'gross income', 'Rating','Total']]
corr=df.corr()
sns.heatmap(corr, cmap='hot', annot=True)
```
